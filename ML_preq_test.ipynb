## Q1 

Total multiplications = n × m × k  
Total additions = n × m × (k-1)  
**Explanation:**  
When we perform a matrix multiplication between a (n, k) and (k, m) matrices, we get an output matrix of nXm matrix. So, the final matrix has n*m elements.  
For finding (i, j)th element in the output matrix, we have to multiply ith-row of the first matrix with the jth-column of the second matrix. Every row of the first matrix has k-elements, so we have k-multiplications for finding (i, j) element and (k-1) additions.  
Total multiplication = n × m × k and addition = n × m × (k-1).


## Q2 

```python
import numpy as np
import time 
# for matrix multiplication, first I'm creating two matrices one is n-rows and k-columns and other one is k-rows and m-columns which has random elements

def create_matrices(n, k, m):
    matrix_1 = np.random.rand(n, k)
    matrix_2 = np.random.rand(k, m)
    return matrix_1, matrix_2

n, k, m = 80, 100, 90
matrix_1, matrix_2 = create_matrices(n, k, m)

# multipling matrices by nested loop 
start_time = time.time()
output_matrix = np.zeros((n,m))
for i in range(n):
    for j in range(m):
        sum_i_j = 0 
        for k1 in range(k):
            sum_i_j += matrix_1[i][k1]*matrix_2[k1][j]
        output_matrix[i][j] = sum_i_j
custom_duration = time.time() - start_time
print("list of list time: ", custom_duration)

start_time_np = time.time()
output_matrix_np = np.dot(matrix_1, matrix_2)
duration_np = time.time()-start_time_np 
print("Numpy time: ",duration_np )
```

When I run on my local machine, I got the following timings: 
- List of List Time: 0.5480690002441406
- NumPy Time: 0.000997304916381836

Here, we can see a significantly difference between numpy inbuilt time and nested loops time. this is because of that the numpy uses parallel operation. 
NumPy takes advantage of multiple processor cores, to execute the program in parallel. This means that instead of processing each element of the matrix sequentially, NumPy can divide the work between multiple CPUs concurrently. So, the overall computation time is much lesser then compared to using nested loops.

```python
# Implementing median by brute force approach
import sys
import numpy as np
import time 
n = 1000
array = np.random.randint(0, 1000, n)
#print(array)
# using nested loop 
start_time_brute = time.time()
taken = {}
second_maxi = -1 
for i in range(n//2+1):
    maxi = -sys.maxsize 
    index = -1 
    for j in range(n):
        if j not in taken:
            if array[j]>maxi:
                maxi = array[j]
                index = j 
    taken[index]= True 
    if i == (n+1)//2-1:
        second_maxi = maxi 
median = maxi 
# if length of array is even then calculating the average of two middle numbers 
if n%2 == 0:
    median = (maxi+second_maxi)/2
duration_brute = time.time()-start_time_brute 
print("Running time of brute force approach: ", duration_brute)
print("Median with Brute force: ", median)


# Calculating the median by numpy funtion
start_time_numpy = time.time()
median_numpy = np.median(array)

duration_numpy = time.time()-start_time_numpy
print("Time of Numpy-median: ", duration_numpy)
print("Median with Numpy funtion:", median_numpy)       
            

# Optimized approach for finding the median 
# first sorting the array and finding middle number 
start_time_sort = time.time()
median_sort = -1 
sorted_array = sorted(array)
if n%2 == 1:
    median_sort = sorted_array[n//2]
else : 
    median_sort = (sorted_array[n//2]+ sorted_array[n//2-1])/2

duration_sort = time.time()-start_time_sort 

print("Running time of optimized approach: ", duration_sort)
print("Median with optimized: ", median_sort)


        
        
```

- time complexity of first approach: O(n*(n/2)) 
- time complexity by using sorting: O(n*log(n))  

- Running time of brute force approach:  0.22100520133972168
- Time of Numpy-median:  0.0009970664978027344
- Running time of optimized approach:  0.0015101432800292969

I got the upper timing for some random examples, I got that numpy.median time is lowest and the second lowest time by sorting array finding median. and the brute force appraoch has largest time.

```python
import numpy as np 
import pandas as pd
import sys 
matrix = np.random.uniform(1, 2, size=(20, 5))

data = pd.DataFrame(matrix ,
                 columns=['a', 'b', 'c', 'd', 'e'])

# Find the column with the highest standard deviation. Find the row with the lowest mean.
maxi_std = -1 
col_no = 'x'

m,n = matrix.shape
for i in data.columns:
    std_of_ith_column = np.std(data[i])
    if maxi_std < std_of_ith_column:
        maxi_std = std_of_ith_column
        col_no = i

print(f"column {col_no}  has Highest Standard Deviation ")

mini_mean = sys.maxsize 
row_no = -1 
for index in data.index:
    row = data.loc[index]
    mean_of_ith_row = np.mean(row)
    if mean_of_ith_row < mini_mean:
        mini_mean = mean_of_ith_row
        row_no = index 

print(f"row {row_no} has lowest Mean  ")

```

```python
import sys 
def fargmin(arr):
    idx = -1 
    mini = sys.maxsize 
    for i in range(len(arr)):
        if arr[i]<mini: 
            idx = i 
            mini = arr[i]
    return idx 

arr = np.random.randint(0,100, 10)
#print(arr)
print(fargmin(arr))
print(np.argmin(arr))
```
